```python
import pandas as pd
import numpy 

sales_data = pd.read_csv('2019 Winter Data Science Intern Challenge Data Set - Sheet1.csv')
```

```python
sales_data
```

```python
#using sample shop to understand value amounts

shop1 = sales_data.query('shop_id == 1')
shop1
```

```python
shop1_avg = shop1['order_amount'].mean()
shop1_avg = round(shop1_avg, 2)
print(f'The mean order for shop id 1 is ${shop1_avg}')
```

```python
#since different stores may have more reasonable AOV
store_avgs = sales_data.groupby(['shop_id'])['order_amount'].mean()
store_avgs
```

```python
#finding stores with outlier sales values 

for store in store_avgs:
    if store > 400: 
        print(store)
```

```python
outlier1 = store_avgs[store_avgs == 235101.49019607843]
outlier2 = store_avgs[store_avgs == 49213.04347826087]
print(outlier1)
print(outlier2)
```

```python
#Looking at the outlier shops to see where the order values are incorrect

shop42 = sales_data.query('shop_id == 42')
shop42
```

```python
#user 607 multiplies total items by 1000 that he strangely makes the purchase at 4 am everyday, instead of deleting 
#the data, I will rectify his mistake rewrite the dataframe 

sales_data['order_amount'] = sales_data['order_amount'].replace([704000], 704)
sales_data['total_items'] = sales_data['total_items'].replace([2000], 2)
```

```python
shop78 = sales_data.query('shop_id == 78')
shop78
```

```python
#all users in this store seem to multiply the sales value by 100, I will update the order amount to reflect a more 
#reasonable purchase

sales_data['order_amount'] = sales_data['order_amount'].replace([25725], 257.25)
sales_data['order_amount'] = sales_data['order_amount'].replace([51450], 514.50)
sales_data['order_amount'] = sales_data['order_amount'].replace([154350], 1543.50)
sales_data['order_amount'] = sales_data['order_amount'].replace([77125], 771.25)
sales_data['order_amount'] = sales_data['order_amount'].replace([102900], 1029.00)
```

```python
avg = sales_data['order_amount'].mean()
avg = round(avg, 2)
print(f'The mean order for all shops is ${avg}')
```
